# OUTLIERS  HANDLING for Ocularsymptomsobservedlately and Symptomsobservingatleasthalfofthetimes

```python
# Assuming 'Ocularsymptomsobservedlately' contains values of type 'object'
data['Ocularsymptomsobservedlately'] = data['Ocularsymptomsobservedlately'].astype(str)

# Function to check if the value is a single digit
def is_single_digit(x):
    return len(x) == 1 and x.isdigit()

# Function to calculate the sum of digits
def calculate_digit_sum(x):
    digits = ''.join(filter(str.isdigit, x))
    return sum(map(int, digits))

# Function to calculate the average digit value
def calculate_average_digit_value(x):
    # Check if it is a single digit
    if is_single_digit(x):
        return int(x)
    
    # Filter out non-integer values
    digits = [int(d) for d in x if d.isdigit()]
    
    # Check if there are multiple digits
    if digits:
        return round(sum(digits) / len(digits))
    else:
        return np.nan

# Iterate through the rows and print the results
for index, row in data.iterrows():
    original_value = row['Ocularsymptomsobservedlately']
    average_value = calculate_average_digit_value(original_value.split('.')[0])
    


    # Update the column with the average value
    data.at[index, 'Ocularsymptomsobservedlately'] = average_value
```
