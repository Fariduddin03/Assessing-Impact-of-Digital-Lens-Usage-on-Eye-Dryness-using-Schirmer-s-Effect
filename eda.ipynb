```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()
```

```python
correlation_matrix = data.corr()

# Exclude self-correlation (correlation of a variable with itself)
correlation_matrix_no_self = correlation_matrix.copy()
n.fill_diagonal(correlation_matrix_no_self.values, 0)

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_no_self, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()
```

```python
pairplot_vars = ['Age', 'Duration', 'workingyears',
                 'hoursspentdailycurricular', 'hoursspentdailynoncurricular',
                 'levelofgadjetwithrespecttoeyes', 'Distancekeptbetweeneyesandgadjet',
                 'Avgnighttimeusageperday', 'Blinkingduringscreenusage',
                 'Difficultyinfocusingafterusingscreens', 'freqquencyofcomplaints',
                 'Severityofcomplaints', 'Complaintsfrequency', 'frequencyofdryeyes',
                 'Schimers1Lefteye', 'Schimers1righteye', 'Schimers2Lefteye',
                 'Schimers2righteye']

# Create pair plot
sns.pairplot(data[pairplot_vars])
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```

```python
output_variables = ['Schimers1Lefteye', 'Schimers1righteye', 'Schimers2Lefteye', 'Schimers2righteye']
for variable in output_variables:
    plt.figure(figsize=(8, 6))
    plt.hist(data[variable], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {variable}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
output_variables = ['Schimers1Lefteye', 'Schimers1righteye', 'Schimers2Lefteye', 'Schimers2righteye']
for variable in output_variables:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=data[variable], color='skyblue')
    plt.title(f'Violin Plot of {variable}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
```
