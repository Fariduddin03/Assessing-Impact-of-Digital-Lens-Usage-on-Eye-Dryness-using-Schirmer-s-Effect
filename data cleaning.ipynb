```python
# Assuming 'Ocularsymptomsobservedlately' contains values of type 'object'
data['Ocularsymptomsobservedlately'] = data['Ocularsymptomsobservedlately'].astype(str)

# Function to check if the value is a single digit
def is_single_digit(x):
    return len(x) == 1 and x.isdigit()

# Function to calculate the sum of digits
def calculate_digit_sum(x):
    digits = ''.join(filter(str.isdigit, x))
    return sum(map(int, digits))

# Function to calculate the average digit value
def calculate_average_digit_value(x):
    # Check if it is a single digit
    if is_single_digit(x):
        return int(x)
    
    # Filter out non-integer values
    digits = [int(d) for d in x if d.isdigit()]
    
    # Check if there are multiple digits
    if digits:
        return round(sum(digits) / len(digits))
    else:
        return np.nan

# Iterate through the rows and print the results
for index, row in data.iterrows():
    original_value = row['Ocularsymptomsobservedlately']
    average_value = calculate_average_digit_value(original_value.split('.')[0])
    


    # Update the column with the average value
    data.at[index, 'Ocularsymptomsobservedlately'] = average_value
```

```python
def categorize_test1(value):
    if value < 5:
        return 'Severe Dry Eyes'
    elif 5 <= value < 10:
        return 'Moderately Dry Eyes'
    elif 10 <= value < 15:
        return 'Mildly Dry Eyes'
    else:
        return 'Normal'

data['Test1Result'] = data['Test1'].apply(categorize_test1)

# Display the DataFrame with the new 'Test1_Category' column
print(data[['Test1', 'Test1Result']])
```

```python
def categorize_test2(value):
    if value < 3:
        return 'Severe Dry Eyes'
    elif 3 <= value < 6:
        return 'Moderately Dry Eyes'
    elif 6 <= value < 10:
        return 'Mildly Dry Eyes'
    else:
        return 'Normal'

data['Test2Result'] = data['Test2'].apply(categorize_test2)

# Display the DataFrame with the new 'Test1_Category' column
print(data[['Test2', 'Test2Result']])
```

```python
def combined_result(row):
    test1 = row['Test1Result']
    test2 = row['Test2Result']
    
    if test1 == test2:
        return test1
    elif test1 == 'Mildly Dry Eyes' and test2 == 'Normal':
        return 'Normal'
    
    elif test1 == 'Moderately Dry Eyes' and test2 == 'Mildly Dry Eyes':
        return 'Mildly Dry Eyes'
    
    elif test1 == 'Severe Dry Eyes' and test2 == 'Moderately Dry Eyes':
        return 'Moderately Dry Eyes'
    
    elif test1 == 'Severe Dry Eyes' and test2 == 'Mildly Dry Eyes':
        return 'Mildly Dry Eyes'
    
    elif test1 == 'Moderately Dry Eyes' and test2 == 'Normal':
        return 'Normal'
    else:
        return 'Indeterminate'

# Apply the combined_result function to each row to determine the combined result
data['Result'] = data.apply(combined_result, axis=1)

# Display the DataFrame with the new 'Result' column
print(data[['Test1Result', 'Test2Result', 'Result']])
```
